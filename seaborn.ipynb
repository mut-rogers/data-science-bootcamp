## Using Seaborn to Visualize data

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Loading the dataset 
dataset = pd.read_csv("euro-daily-hist_1999_2022.csv")
dataset.head()
```

```python
extract = dataset.iloc[:, [0, 1, 4, -2]]
extract.head()
```

```python
# Renaming the columns 
extract.columns = ["Date", "Australian Dollar", "Canadian Dollar", "US Dollar"]
extract.head()
```

```python
sns.lineplot(x=extract.Date, y=extract["US Dollar"], data=extract)
```
