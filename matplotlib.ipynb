```python
# Lets create a simple plot using plt.subplots
x = np.arange(0, np.pi*2, 0.05)
y = np.sin(x) 

fig, ax = plt.subplots() 
ax.plot(x, y, label="Sine", color="green")
ax.legend()
ax.set_title("Simple Sine Curve")
ax.set_ylabel("Sine")
ax.set_xlabel("Angle") 
```

```python
x = np.arange(0, np.pi*2, 0.05) 

fig, ax = plt.subplots(1,2, sharey=True) # The 1 in this case represent the number of rows while 2 represents the number of cols/subplots in that row
# Accessing subplots now using index 
ax[0].plot(x, np.sin(x), label="Sine")
ax[1].plot(x, np.cos(x), label="Cosine") 

# You can then add other features like titles, labels, and legends etc
ax[0].set_xlabel("Angle")
ax[0].set_ylabel("Sin")
ax[0].set_title("Sine Curve")

ax[1].set_ylabel("Cos")
ax[1].set_xlabel("Angle")
ax[1].set_title("Cosine Curve")

# You can also set the overall title of the figure (Entire plot)
fig.suptitle("Comparing Sine and Cosine Curves")
```

```python
# Let's try having more than two plots 
# We shall plot Cosine, Sine, Tangent 

x = np.arange(0, np.pi*2, 0.05) 

fig, ax = plt.subplots(2, 3, sharex=True, sharey=True) # This means we shall have two subplots every rwo giving us a total of 4 subplots 

# We access these subplots using index once more 
# Lets access the first axes, that is, row0 col0 
ax[0, 0].plot(x, np.sin(x), label="Sine", color="green")
ax[0, 0].set_ylabel("sin")
ax[0, 0].set_xlabel("Angle")
ax[0, 0].set_title("Sine Curve")

# Lets access the second axes/plt, that is, row0 col1 
ax[0, 1].plot(x, np.cos(x), label="cos", color="gray")
ax[0, 1].set_ylabel("Cos")
ax[0, 1].set_xlabel("Angle")
ax[0, 1].set_title("Cosine Curve")

# Lets access the third axes/plot, that is row1 col0 
ax[1, 0].plot(x, np.tan(x), label="tan", color="red")
ax[1, 0].set_ylabel("tan")
ax[1, 0].set_xlabel("Angle") 
ax[1, 0].set_title("Tangent Curve") 

# We can also access the fourth axes, that is, row1, col1 
ax[1, 1].plot(x, np.sqrt(x), label="sqrt", color="black")
ax[1, 1].set_ylabel("sqrt")

# We can also set the title of the figure 
fig.suptitle("Various Trigonometric Curves")
```

```python
# Setting up the grid system 
# We can introduce the grid on the figure to improve readability of our plots 
x = np.arange(0, np.pi*2, 0.05) 

fig, ax = plt.subplots(1, 3, sharey=True, figsize=(12,4))
fig.tight_layout()

ax[0].plot(x, np.sin(x), color="blue")
ax[0].set_ylabel("cos")
ax[0].set_xlabel("angle") 
ax[0].grid(visible=True, which="both")
```

```python
x = np.arange(0, np.pi*2, 0.05) 

fig, ax = plt.subplots(1, 3, sharey=True, figsize=(12,4))
fig.tight_layout()

# Setting the grids on the axes
for axes in ax: 
    axes.grid(visible=True, which="both")

ax[0].plot(x, np.sin(x), label="sin", color="blue")
ax[0].plot(x, np.cos(x), label="cos", color="red")
ax[0].set_ylabel("sine")
ax[0].set_xlabel("angle")
ax[0].legend()
```

```python
# Setting the limits on y and x axis
# Matplotlib will set x and y limits by default 
# However, we can explicitlly set these limits using set_ylim and set_xlim
x = np.arange(0, 20, 2)
y = np.square(x)

fig = plt.figure(figsize=(8,4))
fig.tight_layout()

ax=fig.add_axes([0,0,1,1])

ax.plot(x, y, label="square")
ax.grid(visible=True, which="both", color="gray")
ax.set_ylabel("square")
ax.set_xlabel("number")
ax.set_ylim(0,400)
ax.set_xlim(0,20)
ax.set_title("Square of a Number Curve")
ax.legend()
```

## Understanding the Bar Plots
#### - Bar plots are best for representing categorical data. 
#### - The data is represented using rectangles that are proportionate to the values they represent 

```python
lang = ["Python", "C", "c#", "C++", "Java", "JavaScript", "R"]
students = np.random.randint(10, 100, len(lang))

fig = plt.figure(figsize=(10,6))
fig.tight_layout()
ax = fig.add_axes([0,0,1,1]) 

# Plotting the bar 
ax.bar(lang, students)
ax.set_ylim(0,120)
ax.set_xlabel("Programming Language")
ax.set_ylabel("Students")
ax.set_title("Popularity of Programming Languages")

ax.grid(visible=True, which="both")
```

```python
WIDTH: float = 0.25

x = np.arange(0, 4)


fig= plt.figure(figsize=(8,4)) 
ax = fig.add_axes([0,0,1,1])

ax.bar(x, np.random.randint(1, 10, len(x)), color="green", label="CS", width=WIDTH)
ax.bar(x+WIDTH, np.random.randint(1, 10, len(x)), color="red", label="IT", width=WIDTH)
ax.bar(x+WIDTH+WIDTH, np.random.randint(1,10, len(x)), color="blue", label="E&TC", width=WIDTH)

ax.set_xticks([x+0.25 for x in np.arange(4)])
ax.set_xticklabels([2015, 2016, 2017, 2018])
ax.set_ylim(0,10)
ax.grid(visible=True, which="both")

ax.legend(loc="best")
```

```python
WIDTH: float = 0.5 

x = np.arange(0,20,2)

fig = plt.figure(figsize=(12,4))
ax = fig.add_axes([0,0,1,1])

ax.bar(x, np.random.randint(1, 20, len(x)), width=WIDTH, color="red", label="TESLA")
ax.bar(x+WIDTH, np.random.randint(1,20,len(x)), width=WIDTH, color="blue", label="TOYOTA")
ax.bar(x+WIDTH+WIDTH, np.random.randint(1,20,len(x)), width=WIDTH, color="green", label="MAZDA")

ax.set_xticks([x+WIDTH for x in np.arange(len(x))])
ax.set_xticklabels([x+2000 for x in np.arange(10)])
ax.legend(loc="best")
```

## Understanding Histograms
#### - Histograms are accurate representation of the distribution of numerical data
#### - 

```python
a = np.random.randint(15, 100, 100)

fig, ax = plt.subplots(1,1)
ax.hist(a, bins=[0, 25, 75, 100])

ax.set_xticks(np.arange(0, len(a)+1, 25))
ax.set_ylabel("no. of students")
ax.set_xlabel("marks")
ax.set_title("results distribution")
```
